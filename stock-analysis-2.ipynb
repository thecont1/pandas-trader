```python
import pandas as pd
```

```python
df_trader = pd.read_excel("../pandas-trader/stock_trader.xlsx", sheet_name='TRADES')
df_prices = pd.read_excel("../pandas-trader/stock_trader.xlsx", sheet_name='PRICES')
```

```python
df_trader = df_trader.loc[:, 'Broker':'Share Price (₹)']
df_trader.rename({'Broker': 'broker', 'Transaction Date': 'txn_date', 'Exchange': 'exchange', 'Scrip': 'scrip', 'Order Direction': 'order_type', 'Quantity': 'quantity', 'Share Price (₹)': 'price',}, axis=1, inplace=True)

df_trader
```

```python
df_prices = df_prices.loc[:, ['Symbol', 'Company.1', 'Price']]
df_prices.rename({'Symbol': 'symbol', 'Company.1':'company', 'Price': 'price'}, axis=1, inplace=True)

df_prices
```

```python
fields = list(df_trader.keys())

fields
```

```python
row_values = [['Alphanity', '2024-10-25', 'NSE', 'HYUNDAI', 'BUY', 21, 1970.00],
              ['Alphanity', '2024-10-25', 'NSE', 'FACT', 'BUY', '10', 986.21],
              ['Alphanity', '2024-10-25', 'NSE', 'FACT', 'BUY', '10', 986.21],
              ['Alphanity', '2024-10-25', 'NSE', 'FACT', 'BUY', '10', 986.21],
              ['Alphanity', '2024-10-25', 'NSE', 'OLAELEC', 'BUY', '195', 89.01]]

new_rows = pd.DataFrame(row_values, columns=fields)
new_rows
```

```python
df_trader = pd.concat([df_trader, new_rows], ignore_index=True)
df_trader.tail(10)
```

```python
df_trader.drop([258, 259], inplace=True)
df_trader.tail()
```

```python
df_trader.reset_index(drop=True, inplace=True)
df_trader.tail()
```

```python
df_trader['quantity'] = pd.to_numeric(df_trader['quantity'])
df_trader.loc[:, 'trade_value'] = df_trader['quantity'] * df_trader['price']
df_trader.tail()
```

```python
from datetime import date

df_trader['txn_date'] = pd.to_datetime(df_trader['txn_date'])
october_trades = df_trader[df_trader['txn_date'].dt.date >= date(2024,10,1)]
october_trades
```

```python
october_trades = october_trades.merge(df_prices, left_on='scrip', right_on='symbol', sort=True)

fields = {'price_x':'trade_price', 'price_y':'current_price',}
october_trades.rename(fields, axis=1, inplace=True)
```

```python
october_trades['current_worth'] = october_trades['quantity'] * october_trades['current_price']
october_trades['nett_gain'] = october_trades['current_worth'] - october_trades['trade_value']

october_trades
```

```python
required_fields = ['txn_date', 'order_type', 'scrip', 'company', 'quantity', 'trade_price', 'current_price', 'nett_gain']
october_trades = october_trades.loc[:, required_fields][pd.notna(october_trades['current_price'])][october_trades['order_type']=='BUY']
october_trades
```
