```python
import pandas as pd
```

```python
df = pd.read_csv('stock_trades.csv')

df = df.astype({
    'txn_date': 'datetime64[ns]',
    'exchange': 'string',
    'scrip': 'string',
    'buy_qty': 'int64',
    'sell_qty': 'int64',
    'buy_price': 'float64',
    'sell_price': 'float64',
    'brokerage': 'float64',
    'stt': 'float64',
    'other_charges': 'float64',
    'total': 'float64'
})

df['qty'] = df['buy_qty'] - df['sell_qty']
```

```python
df
```

```python
count = 0
scrips = df['scrip'].sort_values().unique().tolist()
for scrip in scrips:
    count += 1
    if count % 10 == 0 or count == len(scrips):
        print(scrip)
    else:
        print(scrip, end=', ')
```

```python
sold_stock = df.groupby('scrip')[['qty', 'total']].sum().sort_values(by='total', ascending=False)
df_profit = sold_stock[(sold_stock['qty'] == 0) & (sold_stock['total'] > 0)]['total']

df_profit
```

```python
df_portfolio = sold_stock[(sold_stock['qty'] != 0)].sort_values(by='total', ascending=True)
df_portfolio['total'] = -round(df_portfolio['total']).astype('int64')

df_portfolio
```

```python
sold_stock = df.groupby('scrip')[['qty', 'total']].sum().sort_values(by='total', ascending=True)
df_loss = sold_stock[(sold_stock['qty'] == 0) & (sold_stock['total'] <= 0)]['total']

df_loss
```
