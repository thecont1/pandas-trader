```python
stocks_data.rename(
    {'Transaction Date': 'txn_date', 
     'Scrip': 'scrip', 
     'Order Direction': 'order_type', 
     'Quantity': 'quantity', 
     'Share Price (₹)': 'price', 
     'Brokerage (₹)': 'charges_brokerage', 
     'ETC (₹)': 'charges_etc', 
     'SEBI+ (₹)': 'charges_sebi', 
     'GST (₹)': 'charges_gst', 
     'STT (₹)': 'charges_stt', 
     'SD (₹)': 'charges_sd'}, 
        axis=1, inplace=True)

stocks_data.drop(columns=['DP ID', 'Client ID', 'Broker', 'Exchange', 'Transaction Time'], axis=1, inplace=True)
```

```python
stocks_data.tail()
```

```python
stocks_data.loc[:, 'charges_brokerage':'charges_sd'].tail()
```

```python
import re
charges = list(filter(lambda f: re.match(r"charges_\w+", f), stocks_data.keys()))
stocks_data[charges].tail()
```

```python
buy_or_sell_factor = stocks_data['order_type'].str.lower().map({'buy': 1, 'sell': -1}).fillna(0)
stocks_data['nett_charges'] = stocks_data[charges].sum(axis=1) * buy_or_sell_factor
stocks_data['nett_charges']
```

```python
stocks_data['nett_trade_value'] = round(stocks_data['quantity'] * stocks_data['price'] + stocks_data['nett_charges'],2)
brief_table = ['txn_date', 'order_type', 'scrip', 'quantity', 'price', 'nett_charges', 'nett_trade_value']
stocks_data[brief_table]
```

```python
buy_transactions = stocks_data[list(map(lambda val: 'buy' in val.lower(), stocks_data['order_type']))][brief_table]
buy_transactions['nett_trade_value'].sum()
```

```python
sell_transactions = stocks_data[list(map(lambda val: 'sell' in val.lower(), stocks_data['order_type']))][brief_table]

pd.Series(
    abs(sell_transactions['nett_trade_value'].sum() - buy_transactions['nett_trade_value'].sum()), 
    index=['Nett Trade (₹):'])
```
